# **Limpieza de la base de datos**

```python
df2=df.drop(columns='Mother Degree')
df2=df2.dropna()
df2 
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
df3=df2.reset_index(drop=True)
df3
```

```python
df3.info()
```

```python
df3.duplicated()
```

```python
df3.duplicated(subset=['Subject_1', 'Subject_3','Subject_5','Subject_8','Subject_10']).sum()
```

```python
df4=df3.drop_duplicates(subset=['Subject_1', 'Subject_3','Subject_5','Subject_8','Subject_10'])
df4
```

```python
df4.duplicated()
```

```python

df4['Student Name'] = df4['Student Name'].replace('bbb', '', regex=True)


print(df4['Student Name'])
```

```python
df4['Student Name'].unique()
```

```python


# Usar pd.to_numeric con errors='coerce' para convertir todo lo que no sea numérico en NaN
df4['Subject_1'] = pd.to_numeric(df4['Subject_1'], errors='coerce')

# Eliminar las filas que contienen NaN
df5 = df4.dropna(subset=['Subject_1'])


df5
```

```python
df5['Subject_1']=df5['Subject_1'].astype(float)
```

```python
df5.columns 
```

```python
conver=['Student Age','Subject_3','Subject_5','Subject_8','Subject_10']

for i in conver:
    df5[i]=pd.to_numeric(df5[i], errors='coerce')

df6=df5.dropna(subset=conver) 

df6
```

```python
conver=['Subject_3','Subject_5','Subject_8','Subject_10']

for j in conver:
    df6[j]=df6[j].astype(float)

df6

```

```python
df6['Student Age']=df6['Student Age'].astype(int)
```

```python
df6.info()
```

```python
df6.isnull().sum()
```

```python
df6.describe()
```

```python

porcentaje_nulos = df6.isnull().mean() * 100


tabla_nulos = pd.DataFrame({
    'Columna': df6.columns,
    'Porcentaje de Nulos': porcentaje_nulos
})


tabla_nulos


```
